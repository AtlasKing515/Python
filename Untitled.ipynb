```python
%matplotlib inline

x=np.linspace(-2,2,101)
y=x**2
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n=int(input("Enter th number of points to draw:"))
x=np.linspace(-1,1,n)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Fuctions cos(2x) 1 sin(2x)")

y=np.cos(2*np.pi*x)
plt.plot(x,y, label="cos(2x)")
y2=np.sin(2*np.pi*x)
plt.plot(x,y2,label="sin(2x)")

plt.legend()
plt.savefig("trigonometric.png")
plt.show()
```
